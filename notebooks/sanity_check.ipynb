```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("../shared_data/olist_order_items_dataset.csv")

print(df.shape)
```

```python
print(df.shape)
```

```python
df["order_id"] = df["order_id"].astype(str).str.strip()

# Count how many times each order_id appears
order_counts = df["order_id"].value_counts()

# Count orders that appear only once
single_orders = (order_counts == 1).sum()

# Count orders that appear more than once
multiple_orders = (order_counts > 1).sum()

# Print results
print("Total rows:", len(df))
print("Unique orders:", df["order_id"].nunique())
print("Orders appearing only once:", single_orders)
print("Orders appearing multiple times:", multiple_orders)
# Visualization
plt.figure(figsize=(6, 4))

categories = ["Single Item Orders", "Multiple Item Orders"]
values = [single_orders, multiple_orders]

plt.bar(categories, values)

plt.title("Order ID Frequency")
plt.ylabel("Number of Orders")

plt.show()
```
