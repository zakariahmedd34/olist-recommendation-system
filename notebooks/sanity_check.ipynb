```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("order_items.csv")

# Count how many times each order_id appears
order_counts = df["order_id"].value_counts()

# Count orders that appear only once
single_orders = (order_counts == 1).sum()

# Count orders that appear more than once
multiple_orders = (order_counts > 1).sum()

# Print results
print(f"Orders appearing only once: {single_orders}")
print(f"Orders appearing multiple times: {multiple_orders}")

# Visualization
plt.figure(figsize=(6, 4))

categories = ["Single Item Orders", "Multiple Item Orders"]
values = [single_orders, multiple_orders]

plt.bar(categories, values)

plt.title("Order ID Frequency")
plt.ylabel("Number of Orders")

plt.show()
```
