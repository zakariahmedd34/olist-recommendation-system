# Compare Full Dataset vs Multi-Item Recommender Results

Run this after the original M5 and the multi-item M5 notebooks.

```
from pathlib import Path
import os

cwd = Path.cwd().resolve()
PROJECT_ROOT = cwd
for candidate in [cwd] + list(cwd.parents):
    if (candidate / "M2_DWH").exists() or (candidate / "olist_dwh.dump").exists():
        PROJECT_ROOT = candidate
        break
os.chdir(PROJECT_ROOT)
print("Project root:", PROJECT_ROOT)
```

Load both comparison tables and combine them.

```
import pandas as pd

full = pd.read_csv("outputs/m5/comparison_table.csv")
multi = pd.read_csv("outputs/m5_multi_item/comparison_table.csv")

full.insert(0, "experiment", "Full Dataset")
multi.insert(0, "experiment", "Multi-Item Orders Only")

comparison = pd.concat([full, multi], ignore_index=True)
comparison.to_csv("outputs/m5_multi_item/full_vs_multi_item_comparison.csv", index=False)
display(comparison)
```

Show only the key metrics used in the discussion.

```
metric_cols = [c for c in ["precision_at_10", "recall_at_10", "hit_rate_at_10", "coverage"] if c in comparison.columns]
display(comparison[["experiment", "system"] + metric_cols])
```
